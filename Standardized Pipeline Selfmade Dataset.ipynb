# Standard Pipeline - Binary model Selfmade

```python
# ----------------------------- Block 1: Last inn datasett, fjern unødvendige kanaler, binær-annotasjon -----------------------------
import os
import pandas as pd
import numpy as np

# For reproducerbarhet
np.random.seed(42)

# 1) Les inn alle CSV-filer
csv_dir = './csv_output'
csv_files = [os.path.join(csv_dir, f) for f in os.listdir(csv_dir) if f.endswith('.csv')]
df = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)

# 2) Standardiser annotation-tekst
df['annotation'] = df['annotation'].str.upper().str.strip()

# 3) Identifiser EEG-kanaler: ta alle kolonner unntatt metadata, fjern AUX og DIR
metadata_cols = ['Person', 'Recording', 'time', 'annotation']
all_channels  = [c for c in df.columns if c not in metadata_cols]
eeg_channels  = [c for c in all_channels if ('AUX' not in c.upper()) and ('DIR' not in c.upper())]

# 4) Filtrer bort uønskede kolonner, behold kun EEG-signaler + metadata
df = df[metadata_cols + eeg_channels].copy()

# 5) Binær-annotasjon: behold kun REST eller IMAGERY-varianter, lag label
df = df[df['annotation'].str.contains('REST|IMAGERY', na=False)].reset_index(drop=True)
df['label'] = np.where(df['annotation'].str.contains('IMAGERY'), 'IMAGERY', 'REST')

# 6) Sjekk datasett og klassedistribusjon
print("DataFrame shape:", df.shape)
print("Antall EEG-kanaler:", len(eeg_channels))
print("Klassefordeling:\n", df['label'].value_counts())
```

```python
# ----------------------------- Block 2: Notch- og båndpass-filtrering (erstat helt) -----------------------------
import pandas as pd
from scipy.signal       import iirnotch, butter, filtfilt

# 1) Sett opp SciPy-filterkoeffisienter
sfreq = 500.0
b_notch, a_notch = iirnotch(50.0, Q=30.0, fs=sfreq)
b_band,  a_band  = butter(4, [1/(sfreq/2), 40/(sfreq/2)], btype='band')

# 2) Hent rådata som numpy-array
data = df[eeg_channels].T.values    # shape (n_ch, n_samples)

# 3) Zero-phase notch + bandpass
seg = filtfilt(b_notch, a_notch, data, axis=1)
seg = filtfilt(b_band,  a_band,  seg,  axis=1)

# 4) Til DataFrame igjen
df_filtered = pd.DataFrame(seg.T, columns=eeg_channels)

# 5) Sjekk at ting stemmer
print(f"Filtered data shape: {df_filtered.shape}  (skal være samme som rådata)")
print("First few samples:", df_filtered.iloc[:5, :5], sep="\n")
```

```python
# ----------------------------- Block 4 (oppdatert): Epoching med gruppelabels -----------------------------
import numpy as np
import pandas as pd

# data_clean: (n_samples, n_channels) fra df_filtered fra Block 2
data_clean = df_filtered[eeg_channels].values    
persons    = df['Person'].values       
recs       = df['Recording'].values    
labels_all = df['label'].values        

# Samme samplingrate og epoch-parametre som live:
sfreq       = 500.0
epoch_samps = int(4 * sfreq)    # 4 s × 500 Hz = 2000 prøver
step        = epoch_samps       # ingen overlap

# Lag epoker, majoritets-label og gruppelabel per epoke
X_epochs, y_epochs, groups = [], [], []
for start in range(0, data_clean.shape[0] - epoch_samps + 1, step):
    seg = data_clean[start:start+epoch_samps]    
    # baseline­korreksjon utført nøyaktig som live
    baseline = seg[:int(0.5*sfreq), :].mean(axis=0, keepdims=True)
    seg_bc = seg - baseline
    
    X_epochs.append(seg_bc.T)  # (n_channels, n_times)
    # flertalls-label for epoken
    lab = pd.Series(labels_all[start:start+epoch_samps]).mode()[0]
    y_epochs.append(lab)
    # gruppe-ID for stratifisert split
    groups.append(f"{persons[start]}__{recs[start]}")

# Konverter til arrays
X_epochs = np.array(X_epochs)    # (n_epochs, n_channels, n_times)
y_epochs = np.array(y_epochs)
groups    = np.array(groups)

# Sjekk at alt ser korrekt ut
print("Epochs shape:", X_epochs.shape)
print("Label-fordeling:\n", pd.Series(y_epochs).value_counts())
print("Antall unike grupper (Person__Recording):", len(np.unique(groups)))
```

```python
# ----------------------------- Block 5 (oppdatert): Group-based split i train/val/test -----------------------------
from sklearn.model_selection import GroupShuffleSplit
import numpy as np
import pandas as pd

# 1) Først: train vs (val+test)
gss1 = GroupShuffleSplit(n_splits=1, test_size=0.30, random_state=42)
train_idx, vt_idx = next(gss1.split(X_epochs, y_epochs, groups))

# 2) Deretter: val vs test på det resterende
gss2 = GroupShuffleSplit(n_splits=1, test_size=0.50, random_state=42)
val_idx_rel, test_idx_rel = next(gss2.split(X_epochs[vt_idx], y_epochs[vt_idx], groups[vt_idx]))
val_idx  = vt_idx[val_idx_rel]
test_idx = vt_idx[test_idx_rel]

# 3) Del opp data
X_train, y_train = X_epochs[train_idx],    y_epochs[train_idx]
X_val,   y_val   = X_epochs[val_idx],      y_epochs[val_idx]
X_test,  y_test  = X_epochs[test_idx],     y_epochs[test_idx]

# 4) Verifiser at ingen grupper overlapper
train_groups = set(groups[train_idx])
val_groups   = set(groups[val_idx])
test_groups  = set(groups[test_idx])
assert train_groups.isdisjoint(val_groups)
assert train_groups.isdisjoint(test_groups)
assert val_groups.isdisjoint(test_groups)
print("Ingen gruppelekkasje mellom train/val/test ✔️")

# 5) Sjekk former og fordeling
print("Train shape:", X_train.shape, " Val shape:", X_val.shape, " Test shape:", X_test.shape)
print("Train distrib:\n", pd.Series(y_train).value_counts())
print("Val distrib:\n",   pd.Series(y_val).value_counts())
print("Test distrib:\n",  pd.Series(y_test).value_counts())
```

```python
# ----------------------------- Block 5 NEW – Balansér KUN treningssettet + evaluer -----------------------------
import numpy as np, pandas as pd
from sklearn.utils           import resample
from pyriemann.estimation    import Covariances
from pyriemann.tangentspace  import TangentSpace
from sklearn.preprocessing   import StandardScaler
from sklearn.pipeline        import Pipeline
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.svm             import SVC
from sklearn.metrics         import (
    balanced_accuracy_score, classification_report,
    confusion_matrix, ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# ---------- 1) Oversampling av TRAIN ---------------------------------------
df_tr = pd.DataFrame({'X': list(X_train), 'y': y_train})
maj_sz = df_tr['y'].value_counts().max()         # størrelsen på majoritets­klassen

rest_bal    = resample(df_tr[df_tr['y'] == 'REST'],
                       replace=True, n_samples=maj_sz, random_state=42)
imagery_bal = resample(df_tr[df_tr['y'] == 'IMAGERY'],
                       replace=True, n_samples=maj_sz, random_state=42)

df_tr_bal   = pd.concat([rest_bal, imagery_bal]).sample(frac=1, random_state=42)
X_tr_bal    = np.stack(df_tr_bal['X'].values)
y_tr_bal    = df_tr_bal['y'].values

print("Etter oversampling (kun train):")
print(pd.Series(y_tr_bal).value_counts(), '\n')
```

```python
# ----------------------------- Block 6–7 (OPPDATERT): Definer, tren og evaluer pipeline -----------------------------
import numpy as np
from pyriemann.estimation      import Covariances
from pyriemann.tangentspace    import TangentSpace
from sklearn.pipeline          import Pipeline
from sklearn.preprocessing     import StandardScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.svm               import SVC
from sklearn.metrics           import balanced_accuracy_score, classification_report, confusion_matrix

from model_utils import CovTransport

pipe = Pipeline([
    ('cov',    Covariances(estimator='oas')),
    ('align',  CovTransport()),
    ('ts',     TangentSpace(metric='riemann')),
    ('scaler', StandardScaler()),
    ('fs',     SelectKBest(mutual_info_classif, k=50)),
    ('svm',    SVC(
                   kernel='rbf',
                   C=1.0,
                   gamma='scale',
                   class_weight=None,
                   probability=True,    # <--- HER
                   random_state=42))
])

# Tren på hele, balanserte datasettet
pipe.fit(X_tr_bal, y_tr_bal)

# 3) Evaluer på Val & Test
for navn, Xf, yf in [('Val', X_val, y_val), ('Test', X_test, y_test)]:
    yp      = pipe.predict(Xf)
    bal_acc = balanced_accuracy_score(yf, yp)
    print(f"\n{navn} balanced accuracy: {bal_acc:.3f}")
    print(classification_report(yf, yp, target_names=['REST','IMAGERY']))
    print(f"{navn} confusion matrix:\n", confusion_matrix(yf, yp, labels=['REST','IMAGERY']))
```

```python
# ----------------------------- Block 8 (fiks): Plot Confusion Matrix for Test-set -----------------------------
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix

# 0) Hvis du bruker end-to-end‐pipeline:
#    y_test_pred = pipe.predict(X_test)
# Eller hvis du bruker forhånds-ekstraherte features:
#    y_test_pred = pipe.predict(X_test_feat)
y_test_pred = pipe.predict(X_test)  

labels = ['REST', 'IMAGERY']

# 1) Skriv ut klassifikasjonsrapport med riktig etikett‐rekkefølge
print("Test Classification Report:")
print(classification_report(
    y_test,
    y_test_pred,
    labels=labels,
    target_names=labels,
    zero_division=0
))

# 2) Beregn og plott forvirringsmatrisen med samme etiketter
cm = confusion_matrix(y_test, y_test_pred, labels=labels)

fig, ax = plt.subplots(figsize=(5, 4))
im = ax.imshow(cm, interpolation='nearest', cmap='Blues')
ax.figure.colorbar(im, ax=ax)
ax.set_xticks([0, 1])
ax.set_xticklabels(labels)
ax.set_yticks([0, 1])
ax.set_yticklabels(labels)
ax.set_xlabel('Predicted label')
ax.set_ylabel('True label')
ax.set_title('Confusion Matrix (Test Set)')

# Annoter hver rute med tall
thresh = cm.max() / 2.0
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, cm[i, j],
                ha='center', va='center',
                color='white' if cm[i, j] > thresh else 'black')

plt.tight_layout()
plt.show()
```

```python
# --- Block 1: Binary EEGNet pipeline with validation ---
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Conv2D, DepthwiseConv2D, SeparableConv2D,
    BatchNormalization, Activation, AveragePooling2D,
    Dropout, Flatten, Dense
)
from tensorflow.keras.constraints import max_norm
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.metrics import (
    balanced_accuracy_score, classification_report,
    confusion_matrix, ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# Map string‐labels til ints
label_map = {'REST':0, 'IMAGERY':1}
y_tr_int = np.array([label_map[y] for y in y_tr_bal])
y_val_int = np.array([label_map[y] for y in y_val])
y_te_int = np.array([label_map[y] for y in y_test])

# Legg på kanal‐dimensjon
X_tr = X_tr_bal[..., np.newaxis]
X_val = X_val[...,     np.newaxis]
X_te = X_test[...,     np.newaxis]

# EEGNet‐bygger
def create_eegnet(nb_classes, Chans, Samples, dropoutRate=0.5):
    inp = Input(shape=(Chans, Samples, 1))
    x = Conv2D(16, (1, 64), padding='same', use_bias=False)(inp)
    x = BatchNormalization()(x)
    x = DepthwiseConv2D((Chans, 1), depth_multiplier=2,
                        depthwise_constraint=max_norm(1.), use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('elu')(x)
    x = AveragePooling2D((1, 4))(x)
    x = Dropout(dropoutRate)(x)
    x = SeparableConv2D(16, (1, 16), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('elu')(x)
    x = AveragePooling2D((1, 8))(x)
    x = Dropout(dropoutRate)(x)
    x = Flatten()(x)
    x = Dense(nb_classes, kernel_constraint=max_norm(0.25))(x)
    out = Activation('softmax')(x)
    return Model(inputs=inp, outputs=out)

# Bygg og kompile
model = create_eegnet(2, X_tr.shape[1], X_tr.shape[2], dropoutRate=0.5)
model.compile(optimizer=Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Callbacks
es   = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
ckpt = ModelCheckpoint('best_eegnet_binary.h5', monitor='val_loss', save_best_only=True)

# Tren med validation_data
model.fit(X_tr, y_tr_int,
          validation_data=(X_val, y_val_int),
          epochs=50, batch_size=16,
          callbacks=[es, ckpt],
          verbose=1)

# Evaluer
y_pred = np.argmax(model.predict(X_te), axis=1)
print("\n=== EEGNet Binary ===")
print("Balanced accuracy:", balanced_accuracy_score(y_te_int, y_pred))
print(classification_report(y_te_int, y_pred,
                            target_names=['REST','IMAGERY'], zero_division=0))
cm = confusion_matrix(y_te_int, y_pred)
disp = ConfusionMatrixDisplay(cm, display_labels=['REST','IMAGERY'])
fig, ax = plt.subplots(figsize=(4,4))
disp.plot(ax=ax, cmap='Blues', colorbar=False)
ax.set_title("EEGNet Binary")
plt.show()
```

# Standard Pipeline - Multiclass model Selfmade

```python
# ----------------------------- Block 1: Last inn datasett, fjern unødvendige kanaler, tre‐klasses annotasjon -----------------------------
import os
import pandas as pd
import numpy as np

# For reproducerbarhet
np.random.seed(42)

# 1) Les inn alle CSV-filer
csv_dir   = './csv_output'
csv_files = [os.path.join(csv_dir, f) for f in os.listdir(csv_dir) if f.endswith('.csv')]
df        = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)

# 2) Standardiser annotation‐tekst
df['annotation'] = df['annotation'].str.upper().str.strip()

# 3) Identifiser EEG‐kanaler
metadata_cols = ['Person','Recording','time','annotation']
all_ch        = [c for c in df.columns if c not in metadata_cols]
eeg_channels  = [c for c in all_ch if ('AUX' not in c.upper()) and ('DIR' not in c.upper())]

# 4) Behold kun EEG‐signal + metadata
df = df[metadata_cols + eeg_channels].copy()

# 5) Tre‐klasses annotasjon: REST, MOVE eller IMAGERY
keep = df['annotation'].str.contains('REST|MOVE|IMAGERY', na=False)
df   = df[keep].reset_index(drop=True)
def map_label(text):
    if 'MOVE'    in text: return 'MOVE'
    if 'IMAGERY' in text: return 'IMAGERY'
    return 'REST'
df['label'] = df['annotation'].apply(map_label)

# 6) Sjekk
print("DataFrame shape:", df.shape)
print("Antall EEG‐kanaler:", len(eeg_channels))
print("Klassefordeling:\n", df['label'].value_counts())
```

```python
import mne

# 1) Opprett MNE RawArray fra DataFrame
sfreq = 500.0  # samplingfrekvens
info  = mne.create_info(ch_names=eeg_channels, sfreq=sfreq, ch_types='eeg')
raw   = mne.io.RawArray(df[eeg_channels].T.values, info)

# 2) Notch-filter ved 50 Hz
raw.notch_filter(freqs=50., fir_design='firwin', verbose=True)

# 3) Båndpass 1–40 Hz
raw.filter(l_freq=1.0, h_freq=40.0, fir_design='firwin', verbose=True)

# 4) Les filtrert signal tilbake til DataFrame
df_filtered = pd.DataFrame(raw.get_data().T, columns=eeg_channels)

# 5) Sjekk
print(raw.info)
print("First filtered values:\n", df_filtered.iloc[:5, :5])
```

```python
import numpy as np
import pandas as pd

data_clean  = raw.get_data().T
persons     = df['Person'].values
recs        = df['Recording'].values
labels_all  = df['label'].values

sfreq       = raw.info['sfreq']
epoch_samps = int(4 * sfreq)
step        = epoch_samps

X_epochs, y_epochs, groups = [], [], []
for start in range(0, len(data_clean) - epoch_samps + 1, step):
    seg      = data_clean[start:start+epoch_samps]
    baseline = seg[:int(0.5*sfreq), :].mean(axis=0, keepdims=True)
    seg_bc   = seg - baseline
    X_epochs.append(seg_bc.T)
    y_epochs.append(pd.Series(labels_all[start:start+epoch_samps]).mode()[0])
    groups.append(f"{persons[start]}__{recs[start]}")

X_epochs = np.array(X_epochs)
y_epochs = np.array(y_epochs)
groups    = np.array(groups)

print("Epochs shape:", X_epochs.shape)
print("Label-fordeling:\n", pd.Series(y_epochs).value_counts())
print("Antall unike grupper (Person__Recording):", len(np.unique(groups)))
```

```python
from sklearn.model_selection import GroupShuffleSplit
import numpy as np
import pandas as pd

# 1) Train vs (val+test)
gss1    = GroupShuffleSplit(n_splits=1, test_size=0.30, random_state=42)
train_idx, vt_idx = next(gss1.split(X_epochs, y_epochs, groups))

# 2) Val vs Test på det resterende
gss2    = GroupShuffleSplit(n_splits=1, test_size=0.50, random_state=42)
val_rel, test_rel = next(gss2.split(X_epochs[vt_idx], y_epochs[vt_idx], groups[vt_idx]))
val_idx, test_idx = vt_idx[val_rel], vt_idx[test_rel]

# 3) Del opp
X_train, y_train = X_epochs[train_idx], y_epochs[train_idx]
X_val,   y_val   = X_epochs[val_idx],   y_epochs[val_idx]
X_test,  y_test  = X_epochs[test_idx],  y_epochs[test_idx]

# 4) Sjekk gruppelekkasje
assert set(groups[train_idx]).isdisjoint(groups[val_idx])
assert set(groups[train_idx]).isdisjoint(groups[test_idx])
assert set(groups[val_idx]).isdisjoint(groups[test_idx])
print("Ingen gruppelekkasje ✔️")

# 5) Form og fordeling
print("Train:", X_train.shape, pd.Series(y_train).value_counts())
print("Val:  ", X_val.shape,   pd.Series(y_val).value_counts())
print("Test: ", X_test.shape,  pd.Series(y_test).value_counts())
```

```python
# ------------------ Block: 1:1:1 balansert oversampling kun på train ------------------
from sklearn.utils import resample
import pandas as pd
import numpy as np

# 1) Lag DataFrame KUN av treningssettet
df_train = pd.DataFrame({'X': list(X_train), 'y': y_train})

# 2) Del opp etter klasse
dfs = {lab: df_train[df_train['y'] == lab]
       for lab in ['REST', 'MOVE', 'IMAGERY']}

# 3) Finn størrelsen på den største klassen i train
n_max = max(len(dfs['REST']), len(dfs['MOVE']), len(dfs['IMAGERY']))

# 4) Oversample hver klasse til n_max
balanced = [
    resample(dfs[lab], replace=True, n_samples=n_max, random_state=42)
    for lab in dfs
]
df_train_bal = pd.concat(balanced).sample(frac=1, random_state=42)

# 5) Pakk ut balansert treningssett
X_train_bal = np.stack(df_train_bal['X'].values)
y_train_bal = df_train_bal['y'].values

# 6) Sjekk at kun train er 1:1:1
print("Train etter 1:1:1-oversampling:\n", pd.Series(y_train_bal).value_counts())
print("Val (uendret):\n", pd.Series(y_val).value_counts())
print("Test (uendret):\n", pd.Series(y_test).value_counts())
```

```python
# ----------------------------- Block X: Multiklasse‐pipeline med CSP, Riemann + SVM -----------------------------
import numpy as np
from pyriemann.estimation      import Covariances
from pyriemann.spatialfilters  import CSP
from pyriemann.tangentspace    import TangentSpace
from sklearn.pipeline          import Pipeline
from sklearn.preprocessing     import StandardScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.svm               import SVC

from model_utils import CovTransport

# Gi MOVE‐klassen litt ekstra vekt
class_weights = {'REST': 1, 'MOVE': 1, 'IMAGERY': 1}

pipe = Pipeline([
    # 1) Estimer SPD‐kovarianser
    ('cov',    Covariances(estimator='oas')),

    # 2) CSP på kovarianser: hent ut 6 komponenter totalt
    ('csp',    CSP(nfilter=6, log=False)),        

    # 3) Fjern person‐bias i tangentrommet
    ('align',  CovTransport()),                     

    # 4) Prosjekter til tangentrom
    ('ts',     TangentSpace(metric='riemann')),      

    # 5) Standardiser + behold alle features (k='all')
    ('scaler', StandardScaler()),
    ('fs',     SelectKBest(mutual_info_classif, k='all')),  

    # 6) SVM med vektet MOVE‐klasse og sannsynligheter
    ('svm',    SVC(
                   kernel='rbf',
                   C=1.0,
                   gamma='scale',
                   class_weight=class_weights,
                   probability=True,
                   random_state=42))
])

# Tren på 1:1:1‐balanserte data
pipe.fit(X_train_bal, y_train_bal)

# Evaluer på Val & Test
from sklearn.metrics import balanced_accuracy_score, classification_report, confusion_matrix

for navn, Xf, yf in [('Val', X_val, y_val), ('Test', X_test, y_test)]:
    yp      = pipe.predict(Xf)
    bal_acc = balanced_accuracy_score(yf, yp)
    print(f"\n{navn} balanced accuracy: {bal_acc:.3f}")
    print(classification_report(
        yf, yp,
        labels=['REST','MOVE','IMAGERY'],
        target_names=['REST','MOVE','IMAGERY']
    ))
    print(f"{navn} confusion matrix:\n",
          confusion_matrix(yf, yp, labels=['REST','MOVE','IMAGERY']))
```

```python
# === Standard Multiclass comparison ===
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    balanced_accuracy_score, classification_report,
    confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.svm import SVC
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from pyriemann.estimation import Covariances
from pyriemann.spatialfilters import CSP
from pyriemann.tangentspace import TangentSpace
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from model_utils import CovTransport

# Forutsetter at X_train_bal, y_train_bal, X_test, y_test er definert

# 1) Felles feature‐ekstraktor
feature_pipe = Pipeline([
    ('cov',    Covariances(estimator='oas')),
    ('csp',    CSP(nfilter=6, log=False)),
    ('align',  CovTransport()),
    ('ts',     TangentSpace(metric='riemann')),
    ('scaler', StandardScaler()),
    ('fs',     SelectKBest(mutual_info_classif, k='all'))
])

# 2) Multiclass‐pipelines
pipelines_mc = {
    'SVM (RBF)': Pipeline([*feature_pipe.steps,
        ('clf', SVC(kernel='rbf', C=1.0, gamma='scale',
                    probability=True, random_state=42))
    ]),
    'LDA': Pipeline([*feature_pipe.steps,
        ('clf', LinearDiscriminantAnalysis(solver='eigen', shrinkage='auto'))
    ]),
    'LogReg (multinomial)': Pipeline([*feature_pipe.steps,
        ('clf', LogisticRegression(multi_class='multinomial', solver='lbfgs',
                                   penalty='l2', C=1.0, max_iter=1000, random_state=42))
    ]),
    'kNN (k=5)': Pipeline([*feature_pipe.steps,
        ('clf', KNeighborsClassifier(n_neighbors=5))
    ]),
    'RandomForest': Pipeline([*feature_pipe.steps,
        ('clf', RandomForestClassifier(n_estimators=100,
                                       class_weight='balanced', random_state=42))
    ]),
    'GradientBoosting': Pipeline([*feature_pipe.steps,
        ('clf', GradientBoostingClassifier(n_estimators=100,
                                           learning_rate=0.1, random_state=42))
    ]),
    'MLP (100-50)': Pipeline([*feature_pipe.steps,
        ('clf', MLPClassifier(hidden_layer_sizes=(100,50),
                              max_iter=300, early_stopping=True, random_state=42))
    ]),
}

results_mc = {}
for name, pipe in pipelines_mc.items():
    pipe.fit(X_train_bal, y_train_bal)
    y_pred = pipe.predict(X_test)
    acc = balanced_accuracy_score(y_test, y_pred)
    results_mc[name] = acc

    print(f"\n=== {name} ===")
    print(f"Balanced accuracy: {acc:.3f}")
    print(classification_report(y_test, y_pred, zero_division=0))

    cm = confusion_matrix(y_test, y_pred, labels=np.unique(y_test))
    disp = ConfusionMatrixDisplay(cm, display_labels=np.unique(y_test))
    fig, ax = plt.subplots(figsize=(4,4))
    disp.plot(ax=ax, cmap='Blues', colorbar=False)
    ax.set_title(name)
    plt.tight_layout()
    plt.show()

print("\n=== Top modeller (multiclass) ===")
for i,(name,acc) in enumerate(sorted(results_mc.items(), key=lambda x: x[1], reverse=True), start=1):
    print(f"{i}. {name}: {acc:.3f}")
```

```python
# --- Block 2: Standard Multiclass EEGNet pipeline med validation ---
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Conv2D, DepthwiseConv2D, SeparableConv2D,
    BatchNormalization, Activation, AveragePooling2D,
    Dropout, Flatten, Dense
)
from tensorflow.keras.constraints import max_norm
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.metrics import (
    balanced_accuracy_score, classification_report,
    confusion_matrix, ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# 1) Map labels til ints
lbls = ['REST','MOVE','IMAGERY']
label_map_mc = {lbl:i for i,lbl in enumerate(lbls)}
y_tr_int_mc  = np.array([label_map_mc[y] for y in y_train])
y_val_int_mc = np.array([label_map_mc[y] for y in y_val])
y_te_int_mc  = np.array([label_map_mc[y] for y in y_test])

# 2) Legg på kanal‐dimensjon (bruk X_train/X_val/X_test fra split)
X_tr_mc  = X_train[..., np.newaxis]
X_val_mc = X_val[...,   np.newaxis]
X_te_mc  = X_test[...,  np.newaxis]

# 3) EEGNet‐bygger
def create_eegnet(nb_classes, Chans, Samples, dropoutRate=0.5):
    inp = Input(shape=(Chans, Samples, 1))
    x = Conv2D(16, (1,64), padding='same', use_bias=False)(inp)
    x = BatchNormalization()(x)
    x = DepthwiseConv2D((Chans,1), depth_multiplier=2,
                        depthwise_constraint=max_norm(1.), use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('elu')(x)
    x = AveragePooling2D((1,4))(x)
    x = Dropout(dropoutRate)(x)
    x = SeparableConv2D(16, (1,16), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('elu')(x)
    x = AveragePooling2D((1,8))(x)
    x = Dropout(dropoutRate)(x)
    x = Flatten()(x)
    x = Dense(nb_classes, kernel_constraint=max_norm(0.25))(x)
    out = Activation('softmax')(x)
    return Model(inputs=inp, outputs=out)

# 4) Bygg og kompile
model_mc = create_eegnet(3, X_tr_mc.shape[1], X_tr_mc.shape[2], dropoutRate=0.5)
model_mc.compile(optimizer=Adam(),
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])

# 5) Callbacks
es_mc   = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
ckpt_mc = ModelCheckpoint('best_eegnet_multiclass.h5', monitor='val_loss', save_best_only=True)

# 6) Tren med validation_data
model_mc.fit(
    X_tr_mc, y_tr_int_mc,
    validation_data=(X_val_mc, y_val_int_mc),
    epochs=25, batch_size=16,
    callbacks=[es_mc, ckpt_mc],
    verbose=1
)

# 7) Evaluer på testsettet
y_pred_mc = np.argmax(model_mc.predict(X_te_mc), axis=1)
print("\n=== EEGNet Multiclass ===")
print("Balanced accuracy:", balanced_accuracy_score(y_te_int_mc, y_pred_mc))
print(classification_report(y_te_int_mc, y_pred_mc, target_names=lbls, zero_division=0))
cm_mc = confusion_matrix(y_te_int_mc, y_pred_mc)
disp_mc = ConfusionMatrixDisplay(cm_mc, display_labels=lbls)
fig, ax = plt.subplots(figsize=(4,4))
disp_mc.plot(ax=ax, cmap='Blues', colorbar=False)
ax.set_title("EEGNet Multiclass")
plt.show()
```

```python
# === Block 1 (oppdatert): Last inn & IIR‐filterbank‐epoke‐ekstrahering ===
import os
import numpy as np
import pandas as pd
from scipy.signal import iirnotch, butter, filtfilt

# 1) Les inn CSV-er og map labels
csv_dir   = './csv_output'
csv_files = [os.path.join(csv_dir,f) for f in os.listdir(csv_dir) if f.endswith('.csv')]
df        = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)
df['annotation'] = df['annotation'].str.upper().str.strip()
mask = df['annotation'].str.contains('REST|MOVE|IMAGERY', na=False)
df = df[mask].reset_index(drop=True)
def map_label(t):
    if 'MOVE'    in t: return 'MOVE'
    if 'IMAGERY' in t: return 'IMAGERY'
    return 'REST'
df['label'] = df['annotation'].apply(map_label)

# 2) Definer EEG-kanaler og rens ikke-numerisk
eeg_channels = [c for c in df.columns
                if c not in ['Person','Recording','time','annotation','label']
                and 'AUX' not in c.upper() and 'DIR' not in c.upper()]
df[eeg_channels] = df[eeg_channels].apply(pd.to_numeric, errors='coerce')
df.dropna(subset=eeg_channels, inplace=True)
df.reset_index(drop=True, inplace=True)

# 3) Bygg rådata‐array og design bredbånds‐IIR
sfreq = 500.0
raw_data = df[eeg_channels].T.values.astype(np.float64)  # (n_ch, n_samples)

# 3a) Notch @50 Hz
b_notch, a_notch = iirnotch(50.0, Q=30.0, fs=sfreq)
data_iir = filtfilt(b_notch, a_notch, raw_data, axis=1)
# 3b) Bandpass 1–100 Hz
b_broad, a_broad = butter(4, [1/(sfreq/2), 100/(sfreq/2)], btype='band')
data_iir = filtfilt(b_broad, a_broad, data_iir, axis=1)
data_iir = data_iir.T  # (n_samples, n_ch)

labels_all = df['label'].values
persons    = df['Person'].values
recs       = df['Recording'].values

# 4) Epoch 4 s non-overlap + baseline + IIR‐filterbank
bands       = {'delta':(1,4),'theta':(4,8),'alpha':(8,12),'beta':(12,30),'gamma':(30,100)}
epoch_samps = int(4 * sfreq)
base_samps  = int(0.5 * sfreq)

X_epochs, y_epochs, groups = [], [], []
for start in range(0, data_iir.shape[0] - epoch_samps + 1, epoch_samps):
    seg = data_iir[start:start+epoch_samps]              # (epoch_samps, n_ch)
    bl  = seg[:base_samps].mean(axis=0, keepdims=True)
    seg = seg - bl                                        # baseline‐korreksjon
    # flertalls‐label og gruppe-ID
    y_epochs.append(pd.Series(labels_all[start:start+epoch_samps]).mode()[0])
    groups .append(f"{persons[start]}__{recs[start]}")
    # IIR‐filterbank
    band_segs = []
    for l,h in bands.values():
        b_band, a_band = butter(4, [l/(sfreq/2), h/(sfreq/2)], btype='band')
        band_seg = filtfilt(b_band, a_band, seg.T, axis=1)  # (n_ch, epoch_samps)
        band_segs.append(band_seg)
    X_epochs.append(np.stack(band_segs, axis=0))         # (n_bands, n_ch, epoch_samps)

X = np.stack(X_epochs, axis=0)    # (n_epochs, n_bands, n_ch, epoch_samps)
y = np.array(y_epochs)
groups = np.array(groups)

print("X shape:", X.shape)
print("Label-fordeling:", pd.Series(y).value_counts())
print("Unike grupper:", len(np.unique(groups)))
```

```python
# === Block 2: Split, balanser & tren Filterbank-CSP→TS→SVM ===
import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.utils import resample
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.svm import SVC
from pyriemann.estimation import Covariances
from pyriemann.spatialfilters import CSP
from pyriemann.tangentspace import TangentSpace
from model_utils import CovTransport
from sklearn.metrics import balanced_accuracy_score, classification_report, confusion_matrix

# 1) Split 60/20/20 stratified på y
sss1 = StratifiedShuffleSplit(n_splits=1, test_size=0.40, random_state=42)
train_i, vt_i = next(sss1.split(X, y))
sss2 = StratifiedShuffleSplit(n_splits=1, test_size=0.50, random_state=42)
val_rel, test_rel = next(sss2.split(X[vt_i], y[vt_i]))
val_i, test_i = vt_i[val_rel], vt_i[test_rel]

# 2) Oversample kun train til 1:1:1
y_tr = y[train_i]
labels, counts = np.unique(y_tr, return_counts=True)
n_max = counts.max()

# samle opp balansert idx
idx_bal = np.concatenate([
    resample(np.where(y_tr==lab)[0], replace=True, n_samples=n_max, random_state=42)
    for lab in labels
])
y_train_bal = y_tr[idx_bal]
X_train_bal = X[train_i][idx_bal]

# 3) BandSelector transformer
from sklearn.base import BaseEstimator, TransformerMixin
class BandSelector(TransformerMixin, BaseEstimator):
    def __init__(self, band_idx): self.band_idx = band_idx
    def fit(self, X, y=None): return self
    def transform(self, X):
        # X: (n_epochs, n_bands, n_ch, n_times)
        return X[:, self.band_idx, :, :]

# 4) FeatureUnion med 5 bånd-bransjer
union = FeatureUnion([
    (name, Pipeline([
        ('select', BandSelector(i)),
        ('cov',    Covariances(estimator='oas')),
        ('csp',    CSP(nfilter=4, log=False)),
        ('align',  CovTransport()),
        ('ts',     TangentSpace(metric='riemann'))
    ]))
    for i,name in enumerate(bands)
])

# 5) Full pipeline
pipe_fb = Pipeline([
    ('union',  union),
    ('scaler', StandardScaler()),
    ('fs',     SelectKBest(mutual_info_classif, k=50)),
    ('svm',    SVC(kernel='rbf', C=1.0, gamma='scale',
                   class_weight='balanced',
                   probability=True,
                   random_state=42))
])

# 6) Tren
pipe_fb.fit(X_train_bal, y_train_bal)

# 7) Evaluer
for name, idx in [('Val', val_i), ('Test', test_i)]:
    y_true = y[idx]
    y_pred = pipe_fb.predict(X[idx])
    print(f"\n{name} balanced accuracy:", balanced_accuracy_score(y_true, y_pred))
    print(classification_report(y_true, y_pred, target_names=['REST','MOVE','IMAGERY']))
    print("Confusion matrix:\n", confusion_matrix(y_true, y_pred, labels=['REST','MOVE','IMAGERY']))
```

```python
# === Block 3: Preprosesser og epoke-ekstraher ny testdata med samme IIR-filterbank ===
import os
import numpy as np
import pandas as pd
from scipy.signal import iirnotch, butter, filtfilt

# 1) Les inn Person5Recording4.csv og map labels
csv_file = './csv_output/Person5Recording4.csv'
df_new   = pd.read_csv(csv_file)
df_new['annotation'] = df_new['annotation'].str.upper().str.strip()
mask = df_new['annotation'].str.contains('REST|MOVE|IMAGERY', na=False)
df_new = df_new[mask].reset_index(drop=True)
df_new['label'] = df_new['annotation'].apply(
    lambda t: 'MOVE' if 'MOVE' in t else ('IMAGERY' if 'IMAGERY' in t else 'REST')
)

# 2) Rens ikke-numerisk (dersom nødvendig)
for ch in eeg_channels:
    df_new[ch] = pd.to_numeric(df_new[ch], errors='coerce')
df_new.dropna(subset=eeg_channels, inplace=True)
df_new.reset_index(drop=True, inplace=True)

# 3) Bygg rådata-array og design IIR-filtre
sfreq    = 500.0
raw_data = df_new[eeg_channels].T.values.astype(np.float64)  # (n_ch, n_samples)

# Notch @50 Hz
b_notch, a_notch = iirnotch(50.0, Q=30.0, fs=sfreq)
data_iir = filtfilt(b_notch, a_notch, raw_data, axis=1)

# Bandpass 1–100 Hz
b_broad, a_broad = butter(4, [1/(sfreq/2), 100/(sfreq/2)], btype='band')
data_iir = filtfilt(b_broad, a_broad, data_iir, axis=1)

# Transponer til (n_samples, n_ch)
data_iir = data_iir.T

labels_new = df_new['label'].values

# 4) Lag epoker + baseline + IIR-filterbank
bands       = {'delta':(1,4),'theta':(4,8),'alpha':(8,12),'beta':(12,30),'gamma':(30,100)}
epoch_samps = int(4 * sfreq)
base_samps  = int(0.5 * sfreq)

epochs_new, y_new = [], []
for start in range(0, data_iir.shape[0] - epoch_samps + 1, epoch_samps):
    seg = data_iir[start:start+epoch_samps]              # (epoch_samps, n_ch)
    bl  = seg[:base_samps].mean(axis=0, keepdims=True)
    seg = seg - bl                                       # baseline-korreksjon

    # majoritets-label
    y_new.append(pd.Series(labels_new[start:start+epoch_samps]).mode()[0])

    # IIR-filterbank
    band_segs = []
    for l, h in bands.values():
        b_band, a_band = butter(4, [l/(sfreq/2), h/(sfreq/2)], btype='band')
        band_seg = filtfilt(b_band, a_band, seg.T, axis=1)  # (n_ch, epoch_samps)
        band_segs.append(band_seg)

    # stack til (n_bands, n_ch, epoch_samps)
    epochs_new.append(np.stack(band_segs, axis=0))

# Final arrays
X_new = np.stack(epochs_new, axis=0)  # (n_epochs_new, n_bands, n_ch, epoch_samps)
y_new = np.array(y_new)

print("Test-set shape:", X_new.shape)
print("Label-counts:\n", pd.Series(y_new).value_counts())
```

```python
# === Block 4: Evaluer ny testdata med pipe_fb ===
from sklearn.metrics import accuracy_score, balanced_accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

# 1) Prediksjoner
y_pred = pipe_fb.predict(X_new)

# 2) Metrics
print("Holdout Person5 – Accuracy: ", accuracy_score(y_new, y_pred))
print("Holdout Person5 – Balanced acc:", balanced_accuracy_score(y_new, y_pred))
print(classification_report(
    y_new, y_pred,
    labels=['REST','MOVE','IMAGERY'],
    target_names=['REST','MOVE','IMAGERY']
))

# 3) Plot forvirringsmatrise
cm = confusion_matrix(y_new, y_pred, labels=['REST','MOVE','IMAGERY'])
fig, ax = plt.subplots(figsize=(5,4))
im = ax.imshow(cm, cmap='Blues', interpolation='nearest')
ax.set_title('Confusion Matrix – Person5Recording4')
ax.set_xticks([0,1,2]); ax.set_xticklabels(['REST','MOVE','IMAGERY'])
ax.set_yticks([0,1,2]); ax.set_yticklabels(['REST','MOVE','IMAGERY'])
thresh = cm.max() / 2.0
for i in range(3):
    for j in range(3):
        ax.text(j, i, cm[i,j],
                ha='center', va='center',
                color='white' if cm[i,j] > thresh else 'black')
fig.colorbar(im, ax=ax)
plt.tight_layout()
plt.show()
```

# Standard Pipeline New Imported Dataset Binary Model

```python
import numpy as np
import mne
from braindecode.datasets import MOABBDataset
from braindecode.preprocessing import preprocess, Preprocessor, exponential_moving_standardize
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.utils import resample
from pyriemann.estimation      import Covariances
from pyriemann.tangentspace    import TangentSpace
from sklearn.preprocessing     import StandardScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.svm               import SVC
from sklearn.pipeline          import Pipeline
from sklearn.metrics           import balanced_accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
from model_utils import CovTransport

# 1) LAST INN FLERE SUBJEKTER & PREPROCESS
mne.set_log_level("ERROR")
SUBJS = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(f"Loading subjects: {SUBJS}")
ds_raw = MOABBDataset("BNCI2014_001", subject_ids=SUBJS)
print("Dataset loaded.")

PRE = [
    Preprocessor('pick_types', eeg=True, meg=False, stim=False),
    Preprocessor(lambda x: x * 1e6),
    Preprocessor('filter', l_freq=1.0, h_freq=40.0),
    Preprocessor('notch_filter', freqs=50.0, fir_design='firwin'),
    Preprocessor(exponential_moving_standardize,
                 factor_new=1e-3, init_block_size=1000),
]
print("Starting preprocessing...")
preprocess(ds_raw, PRE)
print("Preprocessing complete.")

# 2) MANUELL EPOKING 4s UOVERLAPP, 0.5s BASELINE – KUN LEFT/RIGHT
sfreq = ds_raw.datasets[0].raw.info['sfreq']
win   = int(4 * sfreq)
bl    = int(0.5 * sfreq)
int2name = {1: "LEFT", 2: "RIGHT"}

X_epochs, y_epochs = [], []
print("Creating epochs for LEFT/RIGHT...")
for d in ds_raw.datasets:
    data = d.raw.get_data()
    events, _ = mne.events_from_annotations(d.raw)
    for onset, _, code in events:
        if code in int2name and onset + win <= data.shape[1]:
            seg = data[:, onset:onset+win]
            seg = seg - seg[:, :bl].mean(axis=1, keepdims=True)
            X_epochs.append(seg)
            y_epochs.append(int2name[code])
X_epochs = np.stack(X_epochs)
y_epochs = np.array(y_epochs)
print(f"Total epochs created: {X_epochs.shape[0]}")

# 3) GLOBAL STRATIFIED SPLIT 60/20/20
print("Splitting data 60/20/20 (stratified)...")
sss1 = StratifiedShuffleSplit(n_splits=1, test_size=0.40, random_state=42)
train_idx, vt_idx = next(sss1.split(X_epochs, y_epochs))
vt_X, vt_y = X_epochs[vt_idx], y_epochs[vt_idx]
sss2 = StratifiedShuffleSplit(n_splits=1, test_size=0.50, random_state=42)
val_rel, test_rel = next(sss2.split(vt_X, vt_y))
val_idx, test_idx = vt_idx[val_rel], vt_idx[test_rel]

X_train, y_train = X_epochs[train_idx], y_epochs[train_idx]
X_val,   y_val   = X_epochs[val_idx],   y_epochs[val_idx]
X_test,  y_test  = X_epochs[test_idx],  y_epochs[test_idx]
print(f"Train/Val/Test shapes: {X_train.shape}, {X_val.shape}, {X_test.shape}")

# 4) OVERSAMPLING AV TRAIN (1:1)
print("Performing oversampling on training set...")
labs, cnts = np.unique(y_train, return_counts=True)
n_max = cnts.max()
parts = []
for lab in labs:
    sel = y_train == lab
    Xr = resample(X_train[sel], replace=True, n_samples=n_max, random_state=42)
    parts.append((Xr, np.full(n_max, lab)))
X_tr_bal = np.concatenate([p[0] for p in parts], axis=0)
y_tr_bal = np.concatenate([p[1] for p in parts], axis=0)
dist = {lab: int(np.sum(y_tr_bal==lab)) for lab in labs}
print(f"Balanced train distribution: {dist}")

# 5) DEFINER & TRENE BINARY PIPELINE
print("Defining and training pipeline...")
pipe = Pipeline([
    ('cov',    Covariances(estimator='oas')),
    ('align',  CovTransport()),
    ('ts',     TangentSpace(metric='riemann')),
    ('scaler', StandardScaler()),
    ('fs',     SelectKBest(mutual_info_classif, k=50)),
    ('svm',    SVC(kernel='rbf', C=1.0, gamma='scale',
                   probability=True, random_state=42))
])
pipe.fit(X_tr_bal, y_tr_bal)
print("Training complete. Pipeline:")
print(pipe)

# 6) EVALUERING med nytt fargeskjema
from sklearn.metrics import balanced_accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

def eval_bin(name, Xs, ys):
    print(f"\nEvaluating on {name} set:")
    yp = pipe.predict(Xs)
    bal = balanced_accuracy_score(ys, yp)
    print(f"  Balanced accuracy: {bal:.3f}")
    print(classification_report(
        ys, yp,
        labels=["LEFT","RIGHT"],
        target_names=["LEFT","RIGHT"]
    ))

    cm = confusion_matrix(ys, yp, labels=["LEFT","RIGHT"])
    fig, ax = plt.subplots(figsize=(3.5,3.5))

    # Bruk 'Blues' (eller 'cividis') og slå av interpolasjon
    im = ax.imshow(
        cm,
        interpolation='nearest',
        aspect='equal',
        cmap='Blues',
        vmin=0, vmax=cm.max()
    )
    fig.colorbar(im, ax=ax)

    # Tekst i cellene med kontrastfarge
    thresh = cm.max() / 2.0
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            color = 'white' if cm[i, j] > thresh else 'black'
            ax.text(j, i, cm[i, j],
                    ha='center', va='center',
                    color=color)

    ax.set_xticks([0,1])
    ax.set_xticklabels(["LEFT","RIGHT"], rotation=45)
    ax.set_yticks([0,1])
    ax.set_yticklabels(["LEFT","RIGHT"])
    ax.set_xlabel('Predikert')
    ax.set_ylabel('Sann')
    ax.set_title(f"Confusion Matrix – {name}")
    ax.grid(False)           # fjern eventuelt grid
    plt.tight_layout()
    plt.show()


eval_bin("Validation", X_val, y_val)
eval_bin("Test",       X_test, y_test)

```

# Standard Pipeline New Imported Dataset Multiclass Model

```python
import numpy as np
import mne
from braindecode.datasets import MOABBDataset
from braindecode.preprocessing import preprocess, Preprocessor, exponential_moving_standardize
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.utils import resample
from pyriemann.estimation      import Covariances
from pyriemann.spatialfilters  import CSP
from pyriemann.tangentspace    import TangentSpace
from sklearn.preprocessing     import StandardScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.svm               import SVC
from sklearn.pipeline          import Pipeline
from sklearn.metrics           import balanced_accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
from model_utils import CovTransport

# 1) LAST INN ALLE SUBJECTS & PREPROCESS
mne.set_log_level("ERROR")
SUBJS = list(range(1, 10))  # Subjects 1–9
print(f"Loading subjects: {SUBJS}")
ds_raw = MOABBDataset("BNCI2014_001", subject_ids=SUBJS)
print("Dataset loaded.")

PRE = [
    Preprocessor('pick_types', eeg=True, meg=False, stim=False),
    Preprocessor(lambda x: x * 1e6),                  # V → µV
    Preprocessor('filter', l_freq=1.0, h_freq=40.0),  # 1–40 Hz
    Preprocessor('notch_filter', freqs=50.0, fir_design='firwin'),
    Preprocessor(exponential_moving_standardize,
                 factor_new=1e-3, init_block_size=1000),
]
print("Starting preprocessing...")
preprocess(ds_raw, PRE)
print("Preprocessing complete.")

# 2) MANUELL EPOKING 4s UOVERLAPP, 0.5s BASELINE – MULTICLASS
sfreq = ds_raw.datasets[0].raw.info['sfreq']
win   = int(4 * sfreq)
bl    = int(0.5 * sfreq)
int2name = {1:"LEFT", 2:"RIGHT", 3:"FEET", 4:"TONGUE"}

X_epochs, y_epochs = [], []
print("Creating 4s epochs for all classes...")
for d in ds_raw.datasets:
    data, raw = d.raw.get_data(), d.raw
    events, _ = mne.events_from_annotations(raw)
    for onset, _, code in events:
        if code in int2name and onset + win <= data.shape[1]:
            seg = data[:, onset:onset+win]
            seg = seg - seg[:, :bl].mean(axis=1, keepdims=True)
            X_epochs.append(seg)
            y_epochs.append(int2name[code])
X_epochs = np.stack(X_epochs)
y_epochs = np.array(y_epochs)
print(f"Total epochs: {X_epochs.shape[0]}")
unique, counts = np.unique(y_epochs, return_counts=True)
print("Epoch counts per class:", dict(zip(unique, counts)))

# 3) GLOBAL STRATIFIED SPLIT 60/20/20
print("Splitting data 60/20/20 (stratified)...")
sss1 = StratifiedShuffleSplit(n_splits=1, test_size=0.40, random_state=42)
train_idx, vt_idx = next(sss1.split(X_epochs, y_epochs))
vt_X, vt_y = X_epochs[vt_idx], y_epochs[vt_idx]
sss2 = StratifiedShuffleSplit(n_splits=1, test_size=0.50, random_state=42)
val_rel, test_rel = next(sss2.split(vt_X, vt_y))
val_idx, test_idx = vt_idx[val_rel], vt_idx[test_rel]

X_train, y_train = X_epochs[train_idx], y_epochs[train_idx]
X_val,   y_val   = X_epochs[val_idx],   y_epochs[val_idx]
X_test,  y_test  = X_epochs[test_idx],  y_epochs[test_idx]
print("Shapes – Train/Val/Test:", X_train.shape, X_val.shape, X_test.shape)

# 4) OVERSAMPLING AV TRAIN (1:1:1:1)
print("Oversampling training set to balance classes...")
labs, cnts = np.unique(y_train, return_counts=True)
n_max = cnts.max()
parts = []
for lab in labs:
    sel = y_train == lab
    Xr = resample(X_train[sel], replace=True, n_samples=n_max, random_state=42)
    parts.append((Xr, np.full(n_max, lab)))
X_tr_bal = np.concatenate([p[0] for p in parts], axis=0)
y_tr_bal = np.concatenate([p[1] for p in parts], axis=0)
print("Balanced train distribution:", {lab: int(np.sum(y_tr_bal==lab)) for lab in labs})

# 5) DEFINER & TRENE MULTICLASS‐PIPELINE
print("Defining and training multiclass pipeline...")
pipe = Pipeline([
    ('cov',    Covariances(estimator='oas')),
    ('csp',    CSP(nfilter=6, log=False)),
    ('align',  CovTransport()),
    ('ts',     TangentSpace(metric='riemann')),
    ('scaler', StandardScaler()),
    ('fs',     SelectKBest(mutual_info_classif, k='all')),
    ('svm',    SVC(kernel='rbf', C=1.0, gamma='scale',
                   probability=True, random_state=42))
])
pipe.fit(X_tr_bal, y_tr_bal)
print("Training complete.")

# 6) EVALUERING
def eval_mc(name, Xs, ys):
    from matplotlib import cm
    print(f"\nEvaluating on {name} set:")
    yp = pipe.predict(Xs)
    bal = balanced_accuracy_score(ys, yp)
    print(f"  Balanced accuracy: {bal:.3f}")
    print(classification_report(
        ys, yp,
        labels=list(int2name.values()),
        target_names=list(int2name.values())
    ))

    cmat = confusion_matrix(ys, yp, labels=list(int2name.values()))
    fig, ax = plt.subplots(figsize=(4,4))
    # a) ingen interpolasjon, b) 'Blues' gir god kontrast
    im = ax.imshow(
        cmat,
        interpolation='nearest',
        aspect='equal',
        cmap='Blues',
        vmin=0, vmax=cmat.max()
    )
    fig.colorbar(im, ax=ax)

    # Skriv tallene i cellene
    thresh = cmat.max() / 2.0
    for i in range(cmat.shape[0]):
        for j in range(cmat.shape[1]):
            color = 'white' if cmat[i,j] > thresh else 'black'
            ax.text(j, i, cmat[i,j], ha='center', va='center', color=color)

    ax.set_xticks(range(len(int2name)))
    ax.set_xticklabels(int2name.values(), rotation=45)
    ax.set_yticks(range(len(int2name)))
    ax.set_yticklabels(int2name.values())
    ax.set_xlabel('Predikert')
    ax.set_ylabel('Sann')
    ax.set_title(f"Confusion Matrix – {name}")
    ax.grid(False)               # Fjern eventuelle grid-linjer
    plt.tight_layout()
    plt.show()
eval_mc("Validation", X_val, y_val)
eval_mc("Test",       X_test, y_test)
```
